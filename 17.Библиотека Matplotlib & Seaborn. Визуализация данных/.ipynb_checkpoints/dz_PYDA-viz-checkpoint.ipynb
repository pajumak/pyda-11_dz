```python
from pylab import rcParams
rcParams['figure.figsize'] = 10,7

import pandas as pd
import matplotlib.pyplot as plt
```

```python
happiness = pd.read_csv('happiness/2019.csv')
```

```python
happy_country = happiness.loc[:, ['Country or region', 'Score']].sort_values(by='Score', ascending=False).head(15)
```

```python
plt.barh(happy_country['Country or region'], happy_country['Score'])
plt.xlabel('Score')
plt.ylabel('Country or region')
plt.title('Top 15 most happy countries in 2019')
plt.grid()
```

Вывод: самые счастливые страны, почти все европейские

```python
unhappy_country = happiness.loc[:, ['Country or region', 'Score']].sort_values(by='Score', ascending=True).head(15)
unhappy_country = unhappy_country.set_index('Country or region')
unhappy_country.plot(kind='bar')
```

Вывод: самые несчастливые - почти все в Африке

```python
happy_gdp_high = happiness.loc[:, ['Country or region', 'Score', 'GDP per capita']].sort_values(by='GDP per capita', 
                                                                                           ascending=False).head(20)
happy_gdp_high = happy_gdp_high.set_index('Country or region')
happy_gdp_high
```

```python
happy_gdp_low = happiness.loc[:, ['Country or region', 'Score', 'GDP per capita']].sort_values(by='GDP per capita', 
                                                                                           ascending=True).head(20)
happy_gdp_low = happy_gdp_low.set_index('Country or region')
happy_gdp_low
```

```python
happy_gdp = pd.concat([happy_gdp_high, happy_gdp_low])
happy_gdp.plot()
```

Вывод: прослеживается корреляция между уровнем счастья и ВВП 

```python
import seaborn as sns
```

```python
sns_plot = sns.pairplot(
    happiness[['Freedom to make life choices', 'Score']])
sns_plot
```

Вывод: возможность свободного выбора влияет на уровень счастья
