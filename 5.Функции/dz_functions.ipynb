Вам нужно помочь секретарю автоматизировать работу. Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.

код должен быть грамотно декомпозирован (каждая функция отвечает за свою конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется);

```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
```

```python
def command_ds(number):  
    for key , value in directories.items():
        if (str(number) in directories) and (value == []):
            del(directories[key])
            print('Полка удалена. Текущий перечень полок:' , list(directories.keys()))
            return
        elif str(number) in directories:
            print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок:' , list(directories.keys()))
            return
        print('Такой полки не существует. Текущий перечень полок:' , list(directories.keys()))
        return
    
user_command = input('Введите команду:')
if user_command == 'ds':
    command_ds(input('Введите номер полки:'))
```

```python
def command_p(number): 
    for data in documents:
        if str(number) == data['number']:
            print('Владелец документа:' , data['name'])
            return
        else:
            print('Документ не найден в базе')
            return
          
        
def command_s(number):
    for key, value in directories.items():
        for data in value:
            if number == data:
                print('Документ хранится на полке:' , key) 
            else:
                print('Документ не найден в базе')
                return
                
def command_i():
    for data in documents:
        for key, value in directories.items():
            for info in value:
                if data['number'] == info:
                    print('№:', data['number'] , 'тип:' , data['type'] , 'владелец:' , data['name'] , 'полка хранения:' , key)

def command_ads(number):
    if str(number) in directories:
        print('Такая полка уже существует. Текущий перечень полок:' , list(directories.keys()))
    else:
        info = list()
        directories[str(number)] = info
        print('Полка добавлена. Текущий перечень полок:' , list(directories.keys()))

    
def command_ds(place): 
    for key , value in directories.items():
        if (place in key) and (value == []):
            del(directories[key])
            print('Полка удалена. Текущий перечень полок:' , list(directories.keys()))
            return
        elif place in key:
            print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок:' , list(directories.keys()))
            return  
        else:
            print('Такой полки не существует. Текущий перечень полок:' , list(directories.keys())) 
            return

def command_ad(number):   #МНЕ ДОРАБОТАТЬ
    doc_type = input('Введите тип документа:')
    name = input('Введите владельца документа:')
    place = input('Введите полку для хранения:') 
    new_ = {'type' : doc_type , 'number': number , 'name' : name}
    if place not in directories.keys(): 
        print('Такой полки не существует. Добавьте полку командой as. Текущий список документов:' , documents)
    else: 
        if number not in directories.values():
            directories[place].append(number)
            if new_ not in documents:
                documents.append(new_)
                print('Документ добавлен. Текущий список документов:' , documents)
        
def command_d(number):  #МНЕ ДОРАБОТАТЬ
    for data in documents:
        if number == data['number']:
            documents.remove(data)
            print('Документ удален. Текущий список документов:' , documents)
        else:
            print('Документ не найден в базе. Текущий список документов:' , documents)
    
        
                    
user_command = input('Введите команду:')
if user_command == 'p':
    command_p(input('Введите номер документа:'))
elif user_command == 's':
    command_s(input('Введите номер документа:'))
elif user_command == 'i':
    command_i()
elif user_command == 'ads':
    command_ads(input('Введите номер полки:'))
elif user_command == 'ds':
    command_ds(input('Введите номер полки:'))
elif user_command == 'ad':
    command_ad(input('Введите номер документа:'))
elif user_command == 'd':
    command_d(input('Введите номер документа:'))
elif user_command == 'm':
    command_m(input('Введите номер документа:'))
else:
    print('Команда не определена')
```
